```python
import sqlite3
import pandas as pd
from datetime import datetime
import csv
conn = sqlite3.connect("../data/engagements.db")
cursor = conn.cursor()
cursor.execute("SELECT e.comment_text,c.* FROM comment_analysis c LEFT JOIN engagement e  on c.engagement_id = e.id")

rows = cursor.fetchall()
column_names = [description[0] for description in cursor.description]

# Write to CSV
with open("../csv/comment_analysis_export.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(column_names)  # write header
    writer.writerows(rows)         # write data rows

# Close connection
conn.close()
```
